# Question 1

```python
import numpy as np

def checkOutlier(listValues = [], testValue = 1.0):
    stDev = np.std(listValues)
    mean = np.mean(listValues)
    
    if (mean-3*stDev)<testValue<(mean+3*stDev):
        print(testValue, "is not an outlier")
    else:
        print(testValue, "is an outlier")
        
        
#Testing

exampleList = [1,2,3,4,1,2,3,1,5,6,3,24,32,4,1,4]
exampleValue = 7

print(
    np.mean(exampleList)
)
print(
    np.std(exampleList)
)
print(
    checkOutlier(exampleList, 7)
)
print(
    checkOutlier(exampleList, 1000)
)
```

# Question 6

```python
import pandas as pd

abaloneData = pd.read_csv("abalone.csv")
abaloneData
```

```python
#Importing packages
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

#Splitting data
X = abaloneData.iloc[:,0:7]
y = abaloneData['age']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30
)

#decision tree model
model_depth5 = DecisionTreeClassifier(criterion='entropy', max_depth=5)
model_depth5 = model_depth5.fit(X_train, y_train)

model_depth1 = DecisionTreeClassifier(criterion='entropy', max_depth=1)
model_depth1 = model_depth1.fit(X_train, y_train)

#predictions & accuracy
y_predict_depth5 = model_depth5.predict(X_test)
y_predict_depth1 = model_depth1.predict(X_test)

accuracy_depth5 = accuracy_score(y_test,y_predict_depth5)
accuracy_depth1 = accuracy_score(y_test,y_predict_depth1)

print("The accuracy for depth 5:", accuracy_depth5)
print("The accuracy for depth 1:", accuracy_depth1)
```

b) this is a classification task, as we are classifying into a binary output (age = 0 or 1)

c) this is supervised learning, as we have known labelled testing data for age

d) The accuracy for depth 5 is approximately 0.75, whereas it is approximately 0.64 for depth 1

e) the difference lies in the depth the decision tree model undergoes, as increasing depth allows the model to make more partitions. As a result, the partitions increases the fit, and thus increases the accuracy as max depth increases.
